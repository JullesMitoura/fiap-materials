## O que é Pandas?

**Pandas** (*Python Data Analysis Library*) é a biblioteca central para manipulação de **dados tabulares** em Python. Criada por Wes McKinney em 2008 e construída sobre NumPy, tornou-se o padrão da indústria para análise de dados, ciência de dados e engenharia.

### Estruturas de dados principais

| Estrutura | Dimensões | Analogia |
|---|---|---|
| **`Series`** | 1D — índice + valores | Coluna de planilha |
| **`DataFrame`** | 2D — linhas × colunas | Planilha, tabela SQL |

### Por que Pandas?

- **Leitura universal** — CSV, Excel, JSON, SQL, Parquet em uma linha de código  
- **Limpeza de dados** — tratamento de nulos, duplicatas, tipos incorretos  
- **Transformação** — `merge`, `groupby`, `pivot`, `apply`, `melt`  
- **Performance** — operações vetorizadas sobre milhões de linhas sem loops Python  
- **Integração** — base para Matplotlib, Seaborn, Scikit-learn e toda a stack de data science

## <span style="color:#1E90FF;">3.1 Dados Faltantes</span>

````python
# linhas que possuem ao menos um valor faltante
linhas_nan = df[df.isnull().any(axis=1)]

print(f'Total de linhas com NaN: {len(linhas_nan)}')
print(f'Índices: {linhas_nan.index.tolist()}')
linhas_nan
````

````python
# exibir contexto: 2 linhas antes e 2 depois de cada NaN
janela = 2
indices_nan = linhas_nan.index.tolist()

contexto = set()
for i in indices_nan:
    for j in range(max(0, i - janela), min(len(df), i + janela + 1)):
        contexto.add(j)

df.iloc[sorted(contexto)]
````

````python
colunas_nan = [c for c in linhas_nan.columns if linhas_nan[c].isnull().any()]
numericas    = df.select_dtypes(include='number').columns

# opção A — remover as linhas com NaN
df_drop = df.dropna()
print(f'Linhas antes: {len(df)} | Após dropna: {len(df_drop)} | Removidas: {len(df) - len(df_drop)}')
````

````python
# opção B — preencher com a média da coluna
df_mean = df.copy()
df_mean[colunas_nan] = df_mean[colunas_nan].fillna(df[colunas_nan].mean())
print(f'\nValores preenchidos com média:')
print(df_mean.iloc[sorted(contexto)][colunas_nan])
````

````python
# opção C — interpolação linear (melhor para séries temporais)
df_interp = df.copy()
df_interp[numericas] = df_interp[numericas].interpolate(method='linear')
print(f'\nValores preenchidos com interpolação:')
print(df_interp.iloc[sorted(contexto)][colunas_nan])
````

````python
# decisão: aplicar interpolação (dados são séries temporais contínuas)
# a interpolação usa os vizinhos para estimar o valor — mais preciso que a média global
numericas = df.select_dtypes(include='number').columns
df[numericas] = df[numericas].interpolate(method='linear')

print('NaN restantes:', df.isnull().sum().sum())
df.iloc[sorted(contexto)]
````

## <span style="color:#1E90FF;">6. Transformação de Dados</span>

````python
# converter coluna timestamp de string para datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])

# extrair mês e dia da semana como novas colunas
df['mes'] = df['timestamp'].dt.month
df['dia_semana'] = df['timestamp'].dt.day_name()

df[['timestamp', 'mes', 'dia_semana']].head()
````

````python
# criar coluna derivada: diferença de temperatura do glicol (entrada - saída)
df['delta_glicol'] = df['t_glycol_int'] - df['t_glycol_out']

df[['timestamp', 't_glycol_int', 't_glycol_out', 'delta_glicol']].head()
````

````python
# renomear colunas para nomes mais legíveis
df_renamed = df.rename(columns={
    't_water_in':   'temp_agua_entrada',
    't_glycol_int': 'temp_glicol_entrada',
    't_glycol_out': 'temp_glicol_saida',
    't_water_out':  'temp_agua_saida',
})

df_renamed.columns.tolist()
````

---
## <span style="color:#1E90FF;">7. Agrupamento com GroupBy</span>

````python
# média de eficiência e temperatura da água por mês
df.groupby('mes')[['efficiency', 't_water_in']].mean().round(2)
````

````python
# agg() aplica múltiplas funções ao mesmo tempo
df.groupby('mes')['efficiency'].agg(['mean', 'min', 'max', 'std']).round(3)
````
